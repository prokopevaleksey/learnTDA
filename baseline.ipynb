```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from sklearn.cluster import *
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
from collections import Counter
import networkx as nx
```

```python
def generate_data(num_points=1000,n_dims=3,topology='sphere'):
    if topology == 'sphere':
        data = np.random.randn(num_points,n_dims)
        data = data/np.linalg.norm(data,axis=1)[:,None]
    return data
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(16,16))
data = generate_data()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data.T[0],data.T[1],data.T[2])
plt.show()
```

```python
def filtration(data,f=lambda x: x.T[0]):
    return f(data)
```

```python
filtered = filtration(data)
```

```python
plt.figure(figsize=(16,8))
plt.scatter(data.T[0],data.T[1],c=filtered,marker='.')
```

```python
def binning(filtration_values,n_intervals=10,overlap=0.1):
    min_value = np.min(filtration_values)
    max_value = np.max(filtration_values)
    borders = np.linspace(min_value,max_value,n_intervals+1)
    intervals = []
    indexes = []
    for low_bound, high_bound in zip(borders, borders[1:]):
        low_bound = low_bound-overlap*abs(high_bound-low_bound)
        high_bound = high_bound+overlap*abs(high_bound-low_bound)
        intervals.append((low_bound,high_bound))
        indexes.append(np.where((filtration_values>low_bound)&(filtration_values<high_bound))[0])    
    return intervals,indexes
```

```python
bins, indexes = binning(filtered)
```

```python
plt.figure(figsize=(16,8))
colors = np.random.permutation(cm.rainbow(np.linspace(0,1,len(bins))))
colors[:,3] = 0.5
for i,idx in enumerate(indexes):
    a,b = bins[i]
    plt.scatter(data[idx].T[0],data[idx].T[1],c=colors[i],marker='.')
    plt.vlines([a,b],np.min(data.T[1]),np.max(data.T[1]),colors=colors[i],linestyles='dashed')
```

```python
def cluster(X):
    Z = linkage(X, 'ward')
    max_dist = np.max(Z[:,2])
    min_dist = np.min(Z[:,2])
    n_clusters = []
    thresholds = np.linspace(min_dist,max_dist,1000)
    for x in thresholds:
        clusters = fcluster(Z,x,criterion='distance')
        n_clusters.append(len(set(clusters)))
    n,_ = sorted(Counter(n_clusters).items(),key = lambda t: t[1],reverse=True)[0]
    clusters = fcluster(Z,n,'maxclust')
    cluster2id = [np.where(clusters==i+1)[0] for i in range(len(set(clusters)))]
            
    return cluster2id
```

```python
nodes = []
colors = cm.rainbow(np.linspace(0,1,len(bins)))
node_colors = []
for i,subset in enumerate(indexes):
    nodes.extend(cluster(data[subset]))
    node_colors.extend([colors[i]]*len(cluster(data[subset])))
```

```python
def intersection(x,y):
    return len(np.intersect1d(x,y))
```

```python
distance_matrix = np.zeros([len(nodes),len(nodes)])
for i in range(len(nodes)):
    for j in range(len(nodes)):
        if i != j:
            distance_matrix[i,j] = intersection(nodes[i],nodes[j])
```

```python
plt.imshow(distance_matrix)
```

```python
adjastency_matrix = distance_matrix>np.mean(distance_matrix)
```

```python
plt.imshow(adjastency_matrix)
```

```python
G = nx.from_numpy_matrix(adjastency_matrix)
```

```python
nx.draw(G,node_color=node_colors)
```

```python
pos = [np.mean(data[node],axis=0) for node in nodes]
```

```python
pos = pos/np.linalg.norm(pos,axis=1)[:,None]
```

```python
network_plot_3D(G,30,pos,node_colors)
```

```python
def network_plot_3D(G, angle,pos,node_colors):
    
    n = len(pos)
 
    with plt.style.context(('ggplot')):
        
        fig = plt.figure(figsize=(10,7))
        ax = Axes3D(fig)
        
        for i,(xi,yi,zi) in enumerate(pos):
            ax.scatter(xi, yi, zi, c=node_colors[i],s=100, edgecolors='k', alpha=0.7)
        
        for i,j in enumerate(G.edges()):
 
            x = np.array((pos[j[0]][0], pos[j[1]][0]))
            y = np.array((pos[j[0]][1], pos[j[1]][1]))
            z = np.array((pos[j[0]][2], pos[j[1]][2]))
        
            ax.plot(x, y, z, c='black', alpha=0.5)
    
    ax.view_init(30, angle)
    ax.set_axis_off()
    plt.show()
```
